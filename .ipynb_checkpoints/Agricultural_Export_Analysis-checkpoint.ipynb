# Feature Engineering

```python
# Create new columns for Months
def get_month(date):
    """
    Extracts the month name from a given date string.

    Parameters:
        date (str): A date string.

    Returns:
        Series: A Pandas Series containing the Month names corresponding to the input date.
    """
    # Convert the input date string to a Pandas datetime object with the specified format
    date_column = pd.to_datetime(date)
    # Extract the month names from the datetime column
    month_names = date_column.dt.month_name()


    return month_names

def get_year(date):
    """
    Extracts the year from a given date string.

    Parameters:
        date (str): A date string.

    Returns:
        Series: A Pandas Series containing the years corresponding to the input date.
    """
    # Convert the input date string to a Pandas datetime object with the specified format
    date_column = pd.to_datetime(date)
    # Extract the year from the datetime column
    years = date_column.dt.year

    return years


# Function to define various month and their corresponding season
def get_season(month):
    if month in ['December','January','February']:
        return 'Winter'
    elif month in ['March','April','May']:
        return 'Spring'
    elif month in ['June','July','August']:
        return 'Summer'
    else:
        return 'Fall'
```

```python
# Creating a month column
nig_agric_exp['Month'] = get_month(nig_agric_exp['Date'])
printmd("**Created a `Month` Column**")

# Creating a season column
nig_agric_exp['Season'] = nig_agric_exp['Month'].apply(get_season)
printmd("**Created a `Season` Column**")

# Creating a year Column
nig_agric_exp['Year'] = get_year(nig_agric_exp['Date'])
printmd("**Created a `Year` Column**")
```

# Solutions to the given questions Using Graphical and Non Graphical EDA

## Sales Performance

1. Top-Selling Products:

```python
# Top-selling products
top_selling_products = nig_agric_exp.groupby('Product Name')['Units Sold'].sum().sort_values(ascending=False).head(10)
top_selling_products
```

```python
# plot
top_selling_products.plot(kind='bar', figsize=(12, 6), grid=True, title='Top 10 Selling Products by Units', ylabel="Unit Sold")
```

2. Company with the highest sales revenue:

```python
# Set display option to show complete currency values
pd.set_option('display.float_format', lambda x: '{:,.2f}'.format(x))
```

```python
# Revenue generated by company
company_revenue = nig_agric_exp.groupby('Company')['Export Value'].sum().sort_values(ascending=False)
company_revenue
```

```python
# Plot
company_revenue.plot(kind='bar', figsize=(12, 6), grid=True,title='Company Sales Revenue', ylabel = 'Revenue (Currency)')
plt.gca().ticklabel_format(axis='y', style='plain')  # Display complete currency values on y-axis
```

3. Sales variation across export countries:

- Product sales by country:

```python
# product sales by country
product_sales_by_country = nig_agric_exp.groupby(['Export Country', 'Product Name'])['Units Sold'].sum().unstack()
product_sales_by_country
```

```python
# plot
product_sales_by_country.plot(kind='bar', figsize=(12, 8), title='Product Sales by Export Country', ylabel="Units Sold")
```

- Average Revenue per Country

```python
# average revenue by country
avg_revenue_by_country = nig_agric_exp.groupby('Export Country')['Export Value'].mean()
avg_revenue_by_country
```

```python
# plot
# Plot the average revenue by export country
plt.figure(figsize=(12, 6))
avg_revenue_by_country.plot(kind='bar')
plt.xlabel('Export Country')
plt.ylabel('Average Revenue')
plt.title('Average Revenue by Export Country')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.gca().ticklabel_format(axis='y', style='plain')  # Display complete currency values on y-axis
plt.grid(axis='y')  # Add gridlines for better visualization
plt.show()


```

- Total units sold by country:)

```python
# Total Unit solde by country
units_sold_by_country = nig_agric_exp.groupby('Export Country')['Units Sold'].sum()
units_sold_by_country
```

```python
# plot
units_sold_by_country.plot(kind='bar', figsize=(12, 6), title='Total Units Sold by Export Country', ylabel = "Units Sold")
```

4. Correlation between units sold and profit:

```python
# corr between units sold and profit generated
plt.figure(figsize=(10, 6))
sns.scatterplot(data=nig_agric_exp, x='Units Sold', y='Profit per unit')
plt.title('Correlation between Units Sold and Profit per Unit')
```

## Time-Series Analysis:

1. Sales variation over time:

- Monthly sales:

```python
# nig_agric_exp['Month'] = pd.to_datetime(nig_agric_exp['Date'])
monthly_sales = nig_agric_exp.groupby("Month")['Export Value'].sum()
monthly_sales
```

```python
# Create a line plot for monthly sales trend
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Export Value')
plt.gca().ticklabel_format(axis='y', style='plain')  # Display complete currency values on y-axis
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()
```

- Quarterly Sales:

```python
# Convert 'Month' column to datetime format
nig_agric_exp['Month'] = pd.to_datetime(nig_agric_exp['Date'])

# Group by quarter and sum 'Export Value' for each quarter
quarterly_sales = nig_agric_exp.groupby(pd.Grouper(key='Month', freq='QE'))['Export Value'].sum()
quarterly_sales
```

```python
# Plot the quarterly sales trend
plt.figure(figsize=(12, 6))
quarterly_sales.plot(kind='line', marker='o', linestyle='-')
plt.title('Quarterly Sales Trend')
plt.xlabel('Quarter')
plt.ylabel('Export Value')
plt.gca().ticklabel_format(axis='y', style='plain')  # Display complete currency values on y-axis
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()
```

- Annual Sales:

```python
# Annual sales 
annual_sales = nig_agric_exp.groupby('Year')['Export Value'].sum()
annual_sales
```

```python
# Create a bar plot for annual sales
plt.figure(figsize=(12, 6))
annual_sales.plot(kind='bar')
plt.title('Annual Sales')
plt.xlabel('Year')
plt.ylabel('Export Value')
plt.gca().ticklabel_format(axis='y', style='plain')  # Display complete currency values on y-axis
plt.grid(axis='y')  # Add grid lines only on the y-axis for better readability
plt.show()
```

2. Seasonal trends in sales:

```python
# Seasonal sales trend
seasonal_sales_trend = nig_agric_exp.groupby('Season')['Export Value'].sum()
seasonal_sales_trend
```

```python
# Create a line plot for monthly sales trend
plt.figure(figsize=(12, 6))
plt.plot(seasonal_sales_trend.index, seasonal_sales_trend.values, marker='o', linestyle='-')
plt.title('Seasonal Sales Trend')
plt.xlabel('Season')
plt.ylabel('Export Value')
plt.gca().ticklabel_format(axis='y', style='plain')  # Display complete currency values on y-axis
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()
```

3. Relationship between date and profit margin:

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=nig_agric_exp, x='Date', y='Profit per unit')
plt.title('Relationship between Date and Profit Margin')
```

Cost Analysis:

1. Cost of goods sold (COGS) as a percentage of revenue:

```python
# COGS as a percentage revenue
nig_agric_exp['COGS'] = nig_agric_exp['Export Value'] - nig_agric_exp['Profit per unit'] * nig_agric_exp['Units Sold']
cogs_percentage = (nig_agric_exp['COGS'].sum() / nig_agric_exp['Export Value'].sum()) * 100
printmd(f"COGS as a percentage of revenue: **{cogs_percentage:.2f}%**")
```

2. COGS variation across products:

```python
# COGS Variation by products
cogs_by_product = nig_agric_exp.groupby('Product Name')['COGS'].sum()
cogs_by_product
```

```python
# plot
cogs_by_product.plot(kind='bar', figsize=(12, 6), title='COGS by Product', ylabel="Cost Of Goods Sold(COGS)")
plt.gca().ticklabel_format(axis='y', style='plain')  # Display complete currency values on y-axis
```

Geographic Analysis:

1. Destination ports with the highest export volume:

```python
# Ports with highest export volume
export_volume_by_port = nig_agric_exp.groupby('Destination Port')['Units Sold'].sum().sort_values(ascending=False)
export_volume_by_port
```

```python
# plot
export_volume_by_port.plot(kind='bar', figsize=(12, 6), title='Export Volume by Destination Port', ylabel = "Unit Sold")
```

2. Common transportation modes for export:

```python
# Mode of transportation for exports
transportation_mode_counts = nig_agric_exp['Transportation Mode'].value_counts()
transportation_mode_counts
```

```python
# plot
transportation_mode_counts.plot(kind='bar', figsize=(12, 6), title='Transportation Mode Usage')
```

3. Destination ports ranked by export value:

```python
# Destination ports by export value
export_value_by_port = nig_agric_exp.groupby('Destination Port')['Export Value'].sum().sort_values(ascending=False)
export_value_by_port
```

```python
# plot
export_value_by_port.plot(kind='bar', figsize=(12, 6), title='Export Value by Destination Port', ylabel = "Export Value")
plt.gca().ticklabel_format(axis='y', style='plain')  # Display complete currency values on y-axis
```

4. Top export product for each port:

```python
# top product by units sold for each port
top_product_by_port = nig_agric_exp.groupby(['Destination Port', 'Product Name'])['Units Sold'].sum().reset_index().sort_values(['Destination Port', 'Units Sold'], ascending=[True, False]).groupby('Destination Port').head(1)
print(top_product_by_port[['Destination Port', 'Product Name', 'Units Sold']])
```

Performance Comparison:

1. Product performance in terms of profit margin:

```python
# Profit margin
profit_margin = (nig_agric_exp['Profit per unit'] / nig_agric_exp['unit_price']) * 100
nig_agric_exp['Profit Margin (%)'] = profit_margin

# Product performance by profit margin
product_profit_margin = nig_agric_exp.groupby('Product Name')['Profit Margin (%)'].mean().sort_values(ascending=False)
product_profit_margin
```

```python
# plot
product_profit_margin.plot(kind='bar', figsize=(12, 6), title='Product Profit Margin', ylabel = "Profit Margin(%)")
```

2. Company performance based on units sold and profit generated:

```python
# Company performance based on unit sold and profit generated
company_performance = nig_agric_exp.groupby('Company')[['Units Sold', 'Profit per unit']].sum().reset_index()
company_performance['Total Profit'] = company_performance['Units Sold'] * company_performance['Profit per unit']
company_performance.sort_values('Total Profit', ascending=False)
```

3. Identifying outliers or underperforming products/companies:

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=nig_agric_exp, x='Units Sold', y='Profit per unit', hue='Product Name')
plt.title('Units Sold vs. Profit per Unit by Product')
```
